```python
plt.figure(figsize=(10,15))

col=df.select_dtypes(include=np.number).columns
for i in range(len(col)):
        
    plt.subplot(4,4,i+1)
    
    sns.boxplot(x=df[col[i]])
       
    plt.xlabel(col[i].replace('_'," "))
    plt.ylabel(' ')
```

```python
cont_data_percentage = df.groupby('Continent')['World Population Percentage'].sum().round(1).sort_values(ascending= False).reset_index()
cont_data_percentage
```

```python
labels= cont_data_percentage['Continent']
labels
```

```python
pp = cont_data_percentage['World Population Percentage']
pp
```

```python
pp.plot.pie(labels= labels, autopct = '%1.2f%%', figsize= (12,6),shadow = True)
plt.title('Continent - Population%')
```

```python
# group by continent and sum the population
continent_total_population = df.groupby("Continent")['2022 Population'].sum().sort_values(ascending=False)
df_continent_total_population_2022 = pd.DataFrame(continent_total_population)
df_continent_total_population_2022['2022 Population Percentage'] = round((df_continent_total_population_2022['2022 Population'] / df_continent_total_population_2022['2022 Population'].sum()) * 100, 1)
df_continent_total_population_2022
```

```python
populationwithincontinent = df.groupby(['Continent'])['2022 Population'].sum().reset_index(name = 'PopulationPerContinentsIn2020')
populationwithincontinent['PopulationInMillions'] =  (populationwithincontinent['PopulationPerContinentsIn2020'].astype(float)/1000000)
```

```python
plt.figure(figsize = (14, 7))
sns.barplot(x = 'Continent', y = 'PopulationInMillions', data=populationwithincontinent)
plt.xticks(rotation = 90)
plt.xlabel('Continent', size = 15)
plt.ylabel('Popultion in Millions', size = 15)
plt.title('Population in Continents \n', size = 20)
# plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
